# Routine Generetion 2.0

## Overview: 

### Steps
1. Import Data from SQL DB
2. Create a new population
3. Select Parents based on Fitness - 
4. Do Crossover to create new Population
5. Do Mutation to bring new genes in the chain
6. Check Termination Criterion
7. Go back to 3 if Termination Criterion not met.

### Methods used
1. Creating an Initial Population - Random
2. Selection of Parents to Create a New Generation - Propotionate selection
3. Next generation - Mixture of Parent and Children at a fixed ratio
4. Crossover - 1 point crossover
5. Mutation - Non-uniform mutation
6. termination criteria - minimal changes in fitness of consecutive generation 

### Parameters to be decide:
1. The size of the population (N), # population is the various routines
2. the set of individuals (I), # Each individual Routine
3. The Num of Genes (Gn) # each individual timeslot
4. The Gene set (G) # 
5. the Parent set (P)
6. the probability of crossover (C), 	# No. of parent-pair that will go through Reproduction
7. the set of children (Ch) 			# children set created after crossover
8. the set of children that will undergo Mutation (Chmi) 
9. the Mutation Factor (Mf), 			# Probability that a gene that will be mutated
// 10. the Mutation Rate (Mr) 				# No of individuals that will be mutated
11. the set of mutated Children (Chmf)
12. the new population (Inew)
13. the fitness function (F)
14. and the termination criterion (T)

```python
import random

def generate_initial_population(population_size, num_slots):
    """Generate an initial population of routines."""
    initial_population = []

    # Generate routines for the entire population
    for _ in range(population_size):
        routine = generate_initial_routine(num_slots)
        initial_population.append(routine)

    return initial_population


def generate_initial_routine(num_slots):
    """Generate an initial routine consisting of slots."""
    initial_routine = []

    # Generate slots for the routine
    for _ in range(num_slots):
        # Generate a random subject and batch for each slot
        subject = generate_random_subject()
        batch = generate_random_batch()
        slot = (subject, batch)
        initial_routine.append(slot)

    return initial_routine


def generate_random_subject():
    """Generate a random subject."""
    subjects = ["Math", "Science", "English", "History", "Geography", "Physics"]
    return random.choice(subjects)


def generate_random_batch():
    """Generate a random batch."""
    batches = ["Batch A", "Batch B", "Batch C", "Batch D"]
    return random.choice(batches)


# Example usage
population_size = 100
num_slots = 35

initial_population = generate_initial_population(population_size, num_slots)
print("Initial Population:")
for idx, routine in enumerate(initial_population, start=1):
    print(f"Routine {idx}: {routine}")
```

```python
def is_valid_assignment(routine, row, col, professor):
    for r in range(row):
        if routine[r][col] == professor:
            return False
    return True


def assign_periods_backtracking(
    professor, start_period, end_period, routine, row, mapclasses, maplabs
):
    if start_period > end_period:
        return True

    for i in range(start_period, end_period + 1):
        if is_valid_assignment(routine, row, i, professor):
            for j in range(
                mapclasses[professor] if professor in mapclasses else maplabs[professor]
            ):
                if routine[row][i + j] == "":
                    routine[row][i + j] = professor
                    if assign_periods_backtracking(
                        professor, i + 1, end_period, routine, row, mapclasses, maplabs
                    ):
                        return True
                    routine[row][i + j] = ""  # Backtrack
        else:
            return False

    return False


arrayclass = [
    [3201, "ProfA", 3],
    [2203, "ProfB", 2],
    # Add more subarrays as needed
]

arraylabs = [
    [3201, "ProfC", 1],
    [2203, "ProfD", 2],
    # Add more subarrays as needed
]

routine = [["" for _ in range(36)] for _ in range(3)]  # Initialize routine array
mapclasses = {}
maplabs = {}

# Fill the mapclasses and maplabs
for subarray in arrayclass:
    mapclasses[subarray[1]] = subarray[2]

for subarray in arraylabs:
    maplabs[subarray[1]] = subarray[2]

# Assign labs to specific periods
for lab in arraylabs:
    assign_periods_backtracking(lab[1], 5, 7, routine, 0, mapclasses, maplabs)
    assign_periods_backtracking(lab[1], 12, 14, routine, 0, mapclasses, maplabs)
    assign_periods_backtracking(lab[1], 19, 21, routine, 0, mapclasses, maplabs)
    assign_periods_backtracking(lab[1], 26, 28, routine, 0, mapclasses, maplabs)
    assign_periods_backtracking(lab[1], 33, 35, routine, 0, mapclasses, maplabs)

for i, year in enumerate([3201, 3202, 3203], start=1):
    for subject in arrayclass:
        assign_periods_backtracking(subject[1], 4, 1, routine, i, mapclasses, maplabs)
        assign_periods_backtracking(subject[1], 11, 8, routine, i, mapclasses, maplabs)
        assign_periods_backtracking(subject[1], 18, 15, routine, i, mapclasses, maplabs)
        assign_periods_backtracking(subject[1], 25, 22, routine, i, mapclasses, maplabs)
        assign_periods_backtracking(subject[1], 32, 29, routine, i, mapclasses, maplabs)

# Printing routine for verification
for i, year in enumerate(routine, start=2):
    print(f"Year {i} routine:")
    for period in range(1, 36):
        print(f"Period {period}: {year[period]}")
```
